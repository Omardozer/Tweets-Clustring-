```python
#Importing important libraries
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
```

```python
#Functions

#jaccard_similarity function to measure the similarity between tweets and centroids
#jaccard Similarity = |A intersection B| / |A union B|
def jaccard_similarity(a, b):
    # convert tweets string to sets so we can use intersection and union methods
    a = set(a) #Centroid
    b = set(b) #Tweet
    # calucate jaccard similarity
    similarity = float(len(a.intersection(b))) / len(a.union(b))
    return similarity
    

#jaccard_Distance function returns the distance between tweets and centroid
#jaccard Distance = 1 - |A intersection B| / |A union B|
def jaccard_Distance(a, b):
  distance = 1 - jaccard_similarity(a, b)
  return distance #As it returns small values that means more similarity[0,1] 
  #if 0 -> They are the same, if 1 -> They are completely different
```

```python
#Data loading

dataset=input("Enter the path of the datafile:") #getting the file path from user
if (dataset== None):
  dataset="bbchealth.txt" #"bbchealth.txt" used by default
df = pd.read_csv(dataset,delimiter='|',names=['Id','Time','Tweets'])#Reading data
```

```python
#Data Cleansing

df.drop(df.columns[[0, 1]], axis=1, inplace = True) #Removing tweet ids and timestamps coloumns
df['Tweets']=df['Tweets'].str.replace('#','') #Removing hashtag symbols
df = df.replace('http://.*', '', regex = True) #Removing URLs
df = df.replace('@\w*' , '', regex = True) #Removing mentions (words starts with'@')
df['Tweets']= df['Tweets'].str.lower() #Converting every word to lowercase
df.head() #Checking data
```

K-Means Clustering Algorithm

```python
def k_means(x,k=3, no_of_iterations=5):
    
    #Randomly choosing Centroids
    centroids=[]
    for n in range(0,k):
      centroids.append(np.random.choice(len(df), replace=False))
     
    #finding the distance between centroids and all the data points
    distances = jaccard_Distance(x, centroids)
     
    #Centroid with the minimum Distance
    points = np.array([np.argmin(i) for i in distances]) #Step 3
     
    #Repeating the above steps for a defined number of iterations
    #Step 4
    for _ in range(no_of_iterations): 
        centroids = []
        for idx in range(k):
            #Updating Centroids by taking mean of Cluster it belongs to
            temp_cent = x[points==idx].mean(axis=0) 
            centroids.append(temp_cent)
 
        centroids = np.vstack(centroids) #Updated Centroids 
         
        distances = jaccard_Distance(x, centroids)
        points = np.array([np.argmin(i) for i in distances])
         
    return points 
```

```python
label = k_means(df,3,5)
 
#Visualize the results
 
u_labels = np.unique(label)
for i in u_labels:
    plt.scatter(df[label == i , 0] , df[label == i , 1] , label = i)
plt.legend()
plt.show()
```
